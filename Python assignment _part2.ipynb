```python
# Task 2 — Cart Operation

def add_item(cart, menu, item_name, quantity):
    if item_name not in menu:
        print(f"  [ERROR] '{item_name}' does not exist in the menu.")
        return
    if not menu[item_name]["available"]:
        print(f"  [UNAVAILABLE] '{item_name}' is currently unavailable.")
        return

    for entry in cart:                        # check if already in cart
        if entry["item"] == item_name:
            entry["quantity"] += quantity     # update quantity in-place
            print(f"  [UPDATED] '{item_name}' quantity → {entry['quantity']}")
            return

    # not found in cart — add fresh entry
    cart.append({
        "item":     item_name,
        "quantity": quantity,
        "price":    menu[item_name]["price"]
    })
    print(f"  [ADDED] '{item_name}' x{quantity}")


def remove_item(cart, item_name):
    for entry in cart:
        if entry["item"] == item_name:
            cart.remove(entry)
            print(f"  [REMOVED] '{item_name}' from cart.")
            return
    print(f"  [NOT FOUND] '{item_name}' is not in the cart.")


def update_quantity(cart, item_name, new_quantity):
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] = new_quantity
            print(f"  [UPDATED] '{item_name}' quantity set to {new_quantity}")
            return
    print(f"  [NOT FOUND] '{item_name}' is not in the cart.")


def print_cart(cart):
    if not cart:
        print("  Cart is empty.")
        return
    for entry in cart:
        line_total = entry["quantity"] * entry["price"]
        print(f"  {entry['item']:<20} x{entry['quantity']}   ₹{line_total:.2f}")


def print_summary(cart):
    subtotal = sum(e["quantity"] * e["price"] for e in cart)
    gst      = subtotal * 0.05
    total    = subtotal + gst

    print("\n" + "=" * 38)
    print("        Order Summary")
    print("=" * 38)
    for e in cart:
        line_total = e["quantity"] * e["price"]
        print(f"  {e['item']:<20} x{e['quantity']}   ₹{line_total:.2f}")
    print("-" * 38)
    print(f"  {'Subtotal:':<26} ₹{subtotal:>6.2f}")
    print(f"  {'GST (5%):':<26} ₹{gst:>6.2f}")
    print(f"  {'Total Payable:':<26} ₹{total:>6.2f}")
    print("=" * 38)


# ── simulation ────────────────────────────────────────────────────────────────

cart = []

print("\nStep 1 — Add Paneer Tikka x2")
add_item(cart, menu, "Paneer Tikka", 2)
print_cart(cart)

print("\nStep 2 — Add Gulab Jamun x1")
add_item(cart, menu, "Gulab Jamun", 1)
print_cart(cart)

print("\nStep 3 — Add Paneer Tikka x1 (should update, not duplicate)")
add_item(cart, menu, "Paneer Tikka", 1)
print_cart(cart)

print("\nStep 4 — Add Mystery Burger (not in menu)")
add_item(cart, menu, "Mystery Burger", 1)
print_cart(cart)

print("\nStep 5 — Add Chicken Wings (unavailable)")
add_item(cart, menu, "Chicken Wings", 1)
print_cart(cart)

print("\nStep 6 — Remove Gulab Jamun")
remove_item(cart, "Gulab Jamun")
print_cart(cart)

print_summary(cart)
```

```python
#Task 3 — Inventory Tracker with Deep Copy
import copy




inventory_backup = copy.deepcopy(inventory)

# Demonstrate deep copy works — change original, backup stays untouched
inventory["Paneer Tikka"]["stock"] = 999
print("After manual change:")
print(f"  inventory['Paneer Tikka']['stock']        = {inventory['Paneer Tikka']['stock']}")
print(f"  inventory_backup['Paneer Tikka']['stock'] = {inventory_backup['Paneer Tikka']['stock']}")

# Restore original state
inventory["Paneer Tikka"]["stock"] = 10
print("\nRestored. Continuing with original inventory.\n")


# ── 2. Order fulfilment — deduct cart quantities 

print("=" * 42)
print("         Order Fulfilment")
print("=" * 42)

for entry in cart:
    name     = entry["item"]
    needed   = entry["quantity"]
    stock    = inventory[name]["stock"]

    if needed > stock:
        print(f"  [WARNING] '{name}': need {needed}, only {stock} in stock. Deducting {stock}.")
        inventory[name]["stock"] = 0
    else:
        inventory[name]["stock"] -= needed
        print(f"  [OK] '{name}': deducted {needed}. Stock remaining: {inventory[name]['stock']}")


# ── 3. Reorder alerts 

print("\n" + "=" * 42)
print("         Reorder Alerts")
print("=" * 42)

alerts_found = False
for name, data in inventory.items():
    if data["stock"] <= data["reorder_level"]:
        alerts_found = True
        print(f"  ⚠  Reorder Alert: {name} — "
              f"Only {data['stock']} unit(s) left "
              f"(reorder level: {data['reorder_level']})")

if not alerts_found:
    print("  All items are sufficiently stocked.")


# ── 4. Compare inventory vs backup 

print("\n" + "=" * 42)
print("  Inventory vs Backup Comparison")
print("=" * 42)
print(f"  {'Item':<20} {'Current':>8} {'Backup':>8}  {'Changed?':>8}")
print("-" * 42)
for name in inventory:
    curr = inventory[name]["stock"]
    back = inventory_backup[name]["stock"]
    flag = "<-- changed" if curr != back else ""
    print(f"  {name:<20} {curr:>8} {back:>8}  {flag}")
```

```python
# ── 1. Total revenue per day 

def print_revenue_table(sales_log):
    print("=" * 44)
    print(f"  {'Date':<14} {'Orders':>6}   {'Revenue':>10}")
    print("-" * 44)
    for date, orders in sales_log.items():
        daily_revenue = sum(order["total"] for order in orders)
        print(f"  {date:<14} {len(orders):>6}   ₹{daily_revenue:>9.2f}")
    print("=" * 44)


# ── 2. Best-selling day 

def best_selling_day(sales_log):
    return max(
        sales_log,
        key=lambda date: sum(o["total"] for o in sales_log[date])
    )


# ── 3. Most ordered item 

def most_ordered_item(sales_log):
    item_counts = {}
    for orders in sales_log.values():
        for order in orders:
            for item in order["items"]:
                item_counts[item] = item_counts.get(item, 0) + 1
    return max(item_counts, key=lambda item: item_counts[item]), item_counts


# ── Initial run 

print("INITIAL SALES LOG\n")
print_revenue_table(sales_log)

best  = best_selling_day(sales_log)
total = sum(o["total"] for o in sales_log[best])
print(f"\n  Best-selling day: {best}  (₹{total:.2f})\n")

top_item, counts = most_ordered_item(sales_log)
print(f"  Most ordered item: '{top_item}' — appears in {counts[top_item]} orders\n")


# ── 4. Add new day and reprint ────────────────────────────────────────────────

sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

print("UPDATED SALES LOG (after adding 2025-01-05)\n")
print_revenue_table(sales_log)

best  = best_selling_day(sales_log)
total = sum(o["total"] for o in sales_log[best])
print(f"\n  Best-selling day: {best}  (₹{total:.2f})\n")


# ── 5. Numbered list of all orders using enumerate ────────────────────────────

print("=" * 62)
print("  All Orders")
print("=" * 62)

counter = 1
for date, orders in sales_log.items():
    for order in orders:
        items_str = ", ".join(order["items"])
        print(f"  {counter:<3} [{date}]  Order #{order['order_id']:<3}"
              f" — ₹{order['total']:>6.2f} — {items_str}")
        counter += 1

```
